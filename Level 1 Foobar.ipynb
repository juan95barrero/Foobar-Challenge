# Foobar Challenge lvl 1

**In each test case, the lists x and y will always contain n non-unique integers where n is at least 1 but never more than 99, and one of the lists will contain an additional unique integer which should be returned by the function. The same n non-unique integers will be present on both lists, but they might appear in a different order like in the examples above. Commander Lambda likes to keep the numbers short, so every worker ID will be between -1000 and 1000.**

**Two solutions**

* *First, i implement the merge sort algorithm in each list, second, i implement a divide and conquer approach to find the number in O(lgn)*

* *The second approach takes both list and find the sum of each of them, and finally substract one value to the other and the result is the missing value*

## First approach

### This approach takes O(n log(n))

```python
# This function sort the arrays in O(n log(n))
def merge_sort(nums):
    def merge(left_list,right_list):
        sorted_list = []
        left_list_index = right_list_index = 0
        left_list_length , right_list_length = len(left_list), len(right_list)
        
        for _ in range(left_list_length + right_list_length):
            if left_list_index < left_list_length and right_list_index < right_list_length:
                
                if left_list[left_list_index] <= right_list[right_list_index]:
                    sorted_list.append(left_list[left_list_index])
                    left_list_index += 1
                else:
                    sorted_list.append(right_list[right_list_index])
                    right_list_index += 1
                    
            elif left_list_index == left_list_length:
                sorted_list.append(right_list[right_list_index])
                right_list_index += 1
                
            elif right_list_index == right_list_length:
                sorted_list.append(left_list[left_list_index])
                left_list_index +=1
                
        return sorted_list
    
    def merge_sorter(nums):
        if len(nums) <= 1:
            return nums
        middle = len(nums) // 2
        left_list = merge_sorter(nums[:middle])
        right_list = merge_sorter(nums[middle:])
        
        return merge(left_list,right_list)
    return merge_sorter(nums)


def finding_number(x, y):
    #Due to always, Y lenght is less than X and previously we reduce the arrays to 3 or less, 
    #we can find the missing number assuming that all the numbers in Y are correct, but one number in X
    #is the target number
    if len(y) == 3 and len(x) == 3:
        i = 0
        while (i<=2):
            if y[i] == x[i]:
                i += 1
                pass
            else:  
                return x[i]
        
    if len(y) == 2 and len(x) == 3:
        i=0
        while(i<=1):
            if y[i] == x[i]:
                i += 1
                pass
            else:
                return x[i]
        
        return x[2]


    # This if compare the lenghts of the X and Y arrays, take the middle index for the shortest array (Y)
    # to find the extra number
    elif len(y) <=len(x):
        middle = len(y) // 2    

        # If the number of both arrays are the same in the middle, we have two options
        if y[middle] == x[middle]:
            
            # In the special case of have x = [..,-4,-3,1,1,..] and y = [...,-4,1,1,....]
            # the middle number can be the same but only if the (middle-1) number are the same
            # we can assure that the extra number is in the second middle, so we take y[middle:] and x[middle:]
            # as new arrays and continue with the search
            if y[middle-1] == x[middle-1]:
                y = y[middle:]
                x = x[middle:]
                return finding_number(x,y)
            
            # In the case that (middle-1) doesn't are the same, we can assure that the extra number is at the fist half
            # so we take the y[:middle+1] and x[:middle+1]
            else:
                y = y[:middle+1]
                x = x[:middle+1]
                return finding_number(x,y)
            
        # In the simple case of the middle number doesn't are the same we assure that the extra number is in the first half
        else:
            y = y[:middle+1]
            x = x[:middle+1]
            return finding_number(x,y)

    # Doing this reassigment, we will ensure that Y always has less lenght than X
    # Always joining in the elif that comparing the lenghts
    else:
        y , x = x , y
        return finding_number(x,y)

x = [-4,-3, 14, 27, 1, 4, 2, 50, 3, 1]
x = merge_sort(x)
y = [-4,-3, 14, 27, 1, 4, 2, 50, 3, 1]
y = merge_sort(y)

a = finding_number(x,y)

print(a)
```

## Second approach

### This solution takes O(n) time

```python
#This function takes both list and calculate the respective sum and find the diference,
#This difference is the missing number
def solution(x, y):
    if len(y) < len(x):
        x, y = y, x
    if len(x) < len(y):
        return(-(sum(x) - sum(y)))
```
